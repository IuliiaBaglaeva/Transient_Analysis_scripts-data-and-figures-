```python
import numpy as np
import glob
import matplotlib.pyplot as plt
import pandas as pd
from tifffile import imread
from matplotlib.colors import ListedColormap
%matplotlib
import statistics
import matplotlib.lines as lines

# Number of slices to read
Nslice = 2538
N = 4883

# Create a colormap
red_channel = np.zeros(256, dtype=np.uint8)
green_channel = np.arange(256, dtype=np.uint8)
blue_channel = np.zeros(256, dtype=np.uint8)
colormap = (red_channel, green_channel, blue_channel)
green_colormap = ListedColormap(np.column_stack([red_channel, green_channel, blue_channel]) / 255)

# Find the OME-TIFF file
ome_file = glob.glob("Cell15ProtocolCaf.ome.tif")
if not ome_file:
    raise FileNotFoundError("OME-TIFF file not found")
filename = ome_file[0]

# Read the image
image = imread(filename)

# Specify the channel to be extracted
channel = 0  # replace with the desired channel number

# Extract single channel data
single_channel_data = image[channel, :Nslice]
rotated_image = np.rot90(single_channel_data)

channel = 1
single_channel_data1 = image[channel, :Nslice]
rotated_image1 = np.rot90(single_channel_data1)

# Read Excel data
data = pd.read_excel('Cell15ProtocolCaf_F_analysis.xlsx', sheet_name='Parameters')
F0 = np.mean(data['Baseline, AU']) 


data_0F = pd.read_excel('Cell15ProtocolCaf_F_analysis.xlsx', sheet_name='Parameters')
t_0F = data_0F['t0, ms'].iloc[:4]
t0_Fp = np.array(t_0F*Nslice/N)


df = pd.read_excel('Cell15ProtocolCaf_F_analysis.xlsx', sheet_name='Transients')
filtered_data_traces = df[df['time, ms'] < N]
F = filtered_data_traces['F, AU']/F0-1
F_A = filtered_data_traces['Approximated F, AU']/F0-1
t_F = filtered_data_traces['time, ms']

data_L = pd.read_excel('Cell15ProtocolCaf_L_analysis.xlsx', sheet_name='Parameters')
L0 = np.mean(data_L['Baseline, um']) 

data_0L = pd.read_excel('Cell15ProtocolCaf_L_analysis.xlsx', sheet_name='Parameters')
t_0L = data_0L['t0, ms'].iloc[:4]
t0_Lp = np.array(t_0L*Nslice/N)

dl = pd.read_excel('Cell15ProtocolCaf_L_analysis.xlsx', sheet_name='Transients')
filtered_data_l = dl[dl['time, ms'] < N]
L = filtered_data_l['L, um']/L0
L_A = filtered_data_l['Approximated L, um']/L0
t_L = filtered_data_l['time, ms']

# Show the single channel image with colormap

fig, axs = plt.subplots(2, 2, figsize=(12, 8))

ax1 = axs[0, 0]
ax2 = axs[1, 0]
ax3 = axs[0, 1]
ax4 = axs[1, 1]


ax1.imshow(rotated_image, cmap=green_colormap, aspect='auto')
ax1.set_title('Calcium transients', fontsize=18)
ax1.axis('off')
for x in t0_Fp:
    ax1.axvline(x=x, color='r', linestyle='--', linewidth=2)

ax2.scatter(t_F, F, color='grey', s=3)
ax2.plot(t_F, F_A, color='black',linewidth=2)
ax2.set_ylabel('$\Delta F/F$', fontsize=18)
ax2.set_xlabel('time, ms', fontsize=18)
ax2.spines['right'].set_visible(False)
ax2.spines['top'].set_visible(False)
ax2.yaxis.tick_left()
ax2.xaxis.tick_bottom()
ax2.set_xlim([0, N])
ax2.set_ylim([-0.5, 4.5])
ax2.tick_params(labelsize = 16)
for t in t_0F:
    ax2.axvline(x=t, color='r', linestyle='--', linewidth=2)


ax3.imshow(rotated_image1, cmap='gray', aspect='auto')
ax3.set_title('Sarcomere shortening', fontsize=18)
ax3.axis('off')
for x in t0_Lp:
    ax3.axvline(x=x, color='r', linestyle='--', linewidth=2)

ax4.scatter(t_L, L, color='grey', s=3)
ax4.plot(t_L, L_A, color='black',linewidth=2)
ax4.set_ylabel('$\Delta L/L_0}$', fontsize=18)
ax4.spines['right'].set_visible(False)
ax4.spines['top'].set_visible(False)
# Only display ticks on the left and bottom spines
ax4.yaxis.tick_left()
ax4.xaxis.tick_bottom()
ax4.set_xlim([0, N])
ax4.set_xlabel('time, ms', fontsize=18)

ymax = ax4.get_ylim()[1]
ymin = ax4.get_ylim()[0]  #  ,    

#   x  t_0L   
for x in t_0L:
    ax4.axvline(x=x, color='r', linestyle='--', linewidth=2)


    
ax4.tick_params(labelsize = 16)
ax4.set_yticks(np.arange(0.8, 1.01, 0.05))
fig.subplots_adjust(hspace=0.01, wspace=0.01)
plt.tight_layout()
plt.show()
```
