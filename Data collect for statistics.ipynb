```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import re
from copy import copy

# Set the path to the directory containing the files
directory_path = r'C:\Users\julba\Running_sedentary_project'
```

```python
folders = ["Sedentary rats","Running rats"]
factor = ["Sedentary","Running"]
tr_type = "L"
table = []
columns = None
for idx,f in enumerate(folders):
    fldrs = os.listdir(f)
    for anim_folder in fldrs:
        print(anim_folder)
        idx_rat = int(re.findall(r'\d+', anim_folder)[0])
        cells = os.listdir(f"{f}/{anim_folder}")
        for cell_folder in cells:
            idx_cell = int(re.findall(r'\d+', cell_folder)[0])
            find_excel = os.listdir(f"{f}/{anim_folder}/{cell_folder}")
            for file in find_excel:
                #create row
                if re.search(f'{tr_type}_analysis.xlsx', file):
                    row_start = [factor[idx], idx_rat,idx_cell]
                    df = pd.read_excel(f"{f}/{anim_folder}/{cell_folder}/{file}")
                    if columns is None:
                        columns = df.columns
                    df_np = df.to_numpy()
                    for i in range(len(df_np)):
                        df_np[i][0] = int(df_np[i][0]) + 1
                        row = copy(row_start)
                        row.extend(df_np[i])
                        table.append(row)
columns_out = ["Factor","Animal id","Cell id","Transient id"]
columns_out.extend(columns[1:])
out_df = pd.DataFrame(table,columns = columns_out)
out_df["Transient id"] = out_df["Transient id"].astype("uint32")
```

```python
#Whole set F
import os
import pandas as pd
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import re
from copy import copy


folders = ["D:/My dissertation/obesity/obesity_project/obese/sedentary","D:/My dissertation/obesity/obesity_project/obese/running"]

factor = [0,1]
tr_type = "F"
table = []
columns = None
for idx,f in enumerate(folders):
    fldrs = os.listdir(f)
    for anim_folder in fldrs:
        print(anim_folder)
        idx_rat = anim_folder
        cells = os.listdir(f"{f}/{anim_folder}")
        for cell_folder in cells:
            idx_cell = int(re.findall(r'\d+', cell_folder)[0])
            find_excel = os.listdir(f"{f}/{anim_folder}/{cell_folder}")
            for file in find_excel:
                #create row
                if re.search(f'{tr_type}_analysis.xlsx', file):
                    row_start = [factor[idx], idx_rat,idx_cell]
                    df = pd.read_excel(f"{f}/{anim_folder}/{cell_folder}/{file}")
                    if columns is None:
                        columns = df.columns
                    df_np = df.to_numpy()

                    # Here's where we limit the number of transients for each cell
                    # We use min function to handle the case where there are less than 10 transients
                    for i in range(min(10, len(df_np))):
                        df_np[i][0] = int(df_np[i][0]) + 1
                        row = copy(row_start)

                        row.extend(df_np[i])
                        table.append(row)
columns_out = ["Factor","Animal id","Cell id","Transient id"]
columns_out.extend(columns[1:])
out_df = pd.DataFrame(table,columns = columns_out)
out_df["Transient id"] = out_df["Transient id"].astype("uint32")

out_df.to_excel("output_file_obese_F.xlsx", index=False)
```
